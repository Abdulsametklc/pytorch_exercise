```python
import torch

X = torch.randn(100,1)
y = 3 * X + 2 + 0.1 * torch.randn(100,1) 
```

```python
import torch.nn as nn

model = nn.Linear(1,1)
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
```

```python
for epoch in range(100):
    y_pred = model(X)
    loss = criterion(y_pred, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")
```
