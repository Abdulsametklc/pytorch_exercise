```python
import torch
import torch.nn as nn
```

```python
X = torch.tensor([[1.0], [2.0], [3.0], [4.0]])
y = torch.tensor([[3.0], [6.0], [9.0], [12.0]])
```

```python
model = nn.Linear(1,1)
```

```python
loss_fn = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
for epoch in range(500):
    y_pred = model(X)
    loss = loss_fn(y_pred, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")
```

```python
for i in range(3,30,1):
    print("Tahmin:", model(torch.tensor([[float(i)]])))
```
