```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

df = pd.read_csv(url)
print(df.head())
```

```python
df["target"] = (df["species"] == "setosa").astype(float)
```

```python
import torch
X = torch.tensor(df[["petal_length"]].values, dtype = torch.float32)
y = torch.tensor(df[["target"]].values, dtype = torch.float32)
```

```python
split = int(len(X) * 0.8)

X_train = X[:split]
y_train = y[:split]

X_test = X[split:]
y_test = y[split:]
```

```python
w = torch.randn(1, requires_grad = True)
b = torch.randn(1, requires_grad = True)
```

```python
lr = 0.01
epochs = 1000

for epoch in range(epochs):
    y_pred = X_train * w + b
    loss = ((y_pred - y_train) ** 2).mean()

    loss.backward()

    with torch.no_grad():
        w -= lr * w.grad
        b -= lr * b.grad
    
    w.grad.zero_()
    b.grad.zero_()
```

```python
with torch.no_grad():
    test_pred = X_test * w + b
    test_loss = ((test_pred - y_test) ** 2).mean()

print("Weight: ", w.item())
print("Bias: ",b.item())
print("Test Loss: ", test_loss.item())
```
