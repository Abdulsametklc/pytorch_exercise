```python
import torch

X = torch.tensor([[2.0], [4.0], [8.0], [10.0]])
y = torch.tensor([[10.0],[16.0],[28.0], [34.0]])
```

```python
w = torch.randn(1, requires_grad = True)
b = torch.randn(1, requires_grad = True)
```

```python
lr = 0.01
epochs = 1000

for epoch in range(epochs):
    y_pred = X * w + b

    loss = ((y_pred - y)**2).mean()

    loss.backward()

    with torch.no_grad():
        w -= lr * w.grad
        b -= lr * b.grad
    w.grad.zero_()
    b.grad.zero_()

print("Weight: ",w.item())
print("Bias: ",b.item())
```
